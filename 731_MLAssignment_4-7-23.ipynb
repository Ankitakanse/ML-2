```python
# Replace the null values as 0.00
# df.replace(np.nan, 0.00, inplace=True)
```

```python
new_df=df.columns[df.isna().any()].tolist()
new_df1=df[new_df].copy()
new_df1.fillna(0.00, inplace=True)
new_df1.sample(5)
```

```python
cl_df=df.copy().drop(new_df1,axis=1)
cl_df.head()
```

```python
mean1=new_df1.mean(numeric_only=True)
df4=new_df1.replace(0.00, mean1)
df4.head()
```

```python
median1=new_df1.median(numeric_only=True)
df5=new_df1.replace(0.00, median1)
```

```python
mode1=new_df1.mode(numeric_only=True)
mode1
```

```python
sns.displot(new_df1['Cash Ratio']) # Before Replacing
sns.displot(df4['Cash Ratio']) # After Repalcing with Mean
sns.displot(df5['Cash Ratio']) # After Repalcing with Median
```

```python
plt.plot(new_df1, color='red') # Before Repalcing
plt.plot(df4, color='blue') # After Repalcing with Mean
plt.plot(df5, color='yellow') # After Repalcing with Median
plt.show() 
```
